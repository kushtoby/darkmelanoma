```python
import matplotlib.pyplot as plt
import cv2
import numpy  as np
from PIL import Image
 
img = cv2.imread('image directory')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img2, cv2.COLOR_RGB2HSV)
plt.imshow(img2)
img3.shape
```

```python
blur = cv2.GaussianBlur(img2, (3,3), 2,2,cv2.BORDER_DEFAULT) 
r, g, b = cv2.split (blur)
```

```python
gauss_r = cv2.adaptiveThreshold(r, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 275, 20)
gauss_g = cv2.adaptiveThreshold(g, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 275, 20)
gauss_b = cv2.adaptiveThreshold(b, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 275, 20)
```

```python
img_cartoon = cv2.merge((gauss_r, gauss_g, gauss_b))
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_cartoon)
```

```python
img1_bg = cv2.cvtColor(img_cartoon, cv2.COLOR_BGR2RGB)
img1_bg = cv2.bitwise_and(img1_bg, img2)
```

```python
images=np.concatenate((img2,img1_bg),axis=1)
plt.imshow(images)
```

```python
img1_bg = cv2.cvtColor(img1_bg, cv2.COLOR_RGB2BGR)
cv2.imwrite('new.jpg',img1_bg)
```
